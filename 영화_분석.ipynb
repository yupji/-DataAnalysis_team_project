```python
# 빈그릇 생성
genre_list = list()

# 장르들을 전부 하나씩 잘라서 빈그릇에 담기
for n in range(200):
    gen = df['genre'][n].split(', ')
    for i in range(len(gen)):
        gen[i] = gen[i].replace(' ', '') # 공백제거
    genre_list.extend(gen)

genre_list
```

```python
# 리스트에 총 몇번 등장했는지 세기

result = Counter(genre_list)

index = list()
cols = list('A')
values = list()

for key in result:
        if result[key] >= 2:
                print(key, result[key])
                index.append(key)
                values.append(result[key])
```

```python
print(index)
print(cols)
print(values)
```

```python
# 장르별로 몇개씩인지 DF 생성
genre_df = pd.DataFrame( values, index, cols )
genre_df
```

```python
# 파이차트로 시각화!
genre_df['A'].plot( kind='pie', figsize=(10, 10), fontsize=8 )

plt.title('영화 장르별 비율')

# 비율
ratio = list()

for n in range(len(index)):
    kk = genre_df['A'][n]
    ratio.append(kk)

plt.pie(ratio, autopct='%.1f%%')

plt.axis('equal')
```

```python
# content_rating 항목 확인
df.content_rating.unique()
```

```python
# 겹치는 항목 바꾸기
def myChange( x ):
  x = x.replace(' ', '')
  x = x.replace('15세관람가', '15세이상관람가')
  x = x.replace('12세관람가', '12세이상관람가')
  x = x.replace('18세관람가', '청소년관람불가')
  return x

tmp = df.content_rating.apply( myChange )
tmp.unique(), len(tmp.unique())
```

```python
df['content_rating'] = tmp
```

```python
# 타입확인
df.dtypes
```

```python
# attendance에서 , 제거
def myChange( x ):
  x = x.replace(',', '')
  return x

tmp = df.attendance.apply( myChange )

# df 컬럼을 tmp로 바꾸기
df['attendance'] = tmp

# attendance의 값을 int 형으로 바꾸기
df['attendance'] = pd.to_numeric(df['attendance'])
```

```python
# 타입확인
df.dtypes
```

```python
plt.figure( figsize=(10,6) )
sns.boxplot( x='content_rating', y='attendance', data=df, palette='Set3' ) 
sns.swarmplot( x='content_rating', y='attendance', data=df, color='0.5')
plt.title('관람등급별 관객수')
plt.show()
```

```python
# 감독별 작품이 몇개인지 DF 생성
content_rating_df = df.groupby('content_rating').agg({
    'content_rating':'count'
    #'show_count':'sum'
})
content_rating_df.rename(columns={'content_rating':'cnt'})
```

```python
# 빈 그릇에 값 하나씩 담는 방식 사용
x = list()
y = list()

# 그릇에 담기
for n in range(len(df.content_rating.unique())):
    xx = content_rating_df.index[n]
    yy = content_rating_df.content_rating[n]
    x.append(xx)
    y.append(yy)

# 그리기
plt.bar(x,y)

# 값 표시 코드
for i, v in enumerate(x):
    plt.text(v, y[i], y[i],
             fontsize = 10, 
             color='blue',
             horizontalalignment='center',
             verticalalignment='bottom')
    
plt.title('관람가등급별 영화수')
plt.show()
```

```python
# 년도 별로 관객수 확인!
# opening_date 값의 연도만 남기고 int형으로 변환
def myChange( x ):
  x = x.split('-')[0]
  return x

tmp = df.opening_date.apply( myChange )

# df 컬럼을 tmp로 바꾸기
df['year'] = tmp

df['year'] = pd.to_numeric(df['year'])
df.head()
```

```python
year_df = df.groupby('year').agg({
    'attendance':'mean'
})
year_df
```

```python
year_df.plot()
plt.xlabel("year")
plt.ylabel("Data")
plt.title('연도별 평균 관객수')
plt.show()
```

```python
# 월 별로 관객수 확인!
# opening_date 값의 연도만 남기고 int형으로 변환
def myChange( x ):
  x = x.split('-')[1]
  return x

tmp = df.opening_date.apply( myChange )

# df 컬럼을 tmp로 바꾸기
df['month'] = tmp

df['month'] = pd.to_numeric(df['month'])
df.head(1)

```
