```python
# 모듈 가져오기
import selenium
from selenium import webdriver as wd
import time
import pandas as pd
```

```python
# driver = wd.Chrome('./chromedriver')
driver = wd.Chrome('C://Users//USER//Desktop//py.project//selenium_level4//./chromedriver')
```

```python
# 접속 후 대기
driver.get('https://www.kobis.or.kr/kobis/business/stat/boxs/findFormerBoxOfficeList.do')

# 강제로 5초 대기(명시적 대기)
time.sleep(1*5)

# 국적 한국 선택
select_country = driver.find_element_by_id('sRepNationCd')
select_country.send_keys( '한국' )

# 조회 버튼 클릭
button = driver.find_element_by_css_selector('div > div.wrap_btn > button')
button.click()
time.sleep(1*2)
```

```python
css_sel = '#td_movie > span > a'
links = driver.find_elements_by_css_selector(css_sel)
len( links )
```

```python
# 빈 그릇 생성
movies_infos = list()

for link in links:
    # 링크클릭
    link.click()
    time.sleep(1*2)
    try:
        # 해당 요소 찾기
        dds = driver.find_elements_by_tag_name('.ovf.cont dd')
        # 정보 추출(장르, 상영시간, 관객등급, 감독)
        genre = dds[3].text.strip().split('|')[2]
        runtime = dds[3].text.strip().split('|')[3]
        content_rating = dds[3].text.strip().split('|')[4]
        director = driver.find_element_by_css_selector('.info.info2 > div > dl > div > dd > a').text.strip()

        # 배우 항목이 여러개여서 항목 수 체크
        length = len(driver.find_elements_by_css_selector('.info.info2 > div > dl > div:nth-child(2) > dd > table:nth-child(1) > tbody > tr > td > a'))

        # 배우를 담을 빈그릇 생성
        actor_list = list()

        # 배우 정보 추출
        for n in range(length):
            actors = driver.find_elements_by_css_selector('.info.info2 > div > dl > div:nth-child(2) > dd > table:nth-child(1) > tbody > tr > td > a')
            act =  actors[n].text.strip().split('(')[0] # 극중 이름 제거
            actor_list.append(act)
        actor = ', '.join(str(s) for s in actor_list) # 여러개의 배우 이름 하나의 문자열로 합치기

        # 그릇에 담기
        movies_infos.append([genre, runtime, content_rating, director, actor])

    except Exception as e:
        print(e)

    # 브라우저 닫기
    driver.execute_script('dtlPrev();')
    time.sleep(1*2)


    
    
movies_infos
    
```

```python
# 데이터프레임으로 변환
movie_df = pd.DataFrame(movies_infos,
columns = ['genre', 'runtime', 'content_rating', 'director', 'actor'])
movie_df
```

```python
# 빈 그릇 생성
distributor_infos = list()

for link in links:
    # 링크클릭
    link.click()
    time.sleep(1*2)
    try:
        distributor = driver.find_element_by_css_selector('dd:nth-child(4) > a').text.strip()

        # 그릇에 담기
        distributor_infos.append([distributor])

    except Exception as e:
        print(e)

    # 브라우저 닫기
    driver.execute_script('dtlPrev();')
    time.sleep(1*2)


    
    
distributor_infos
```

```python
# 데이터프레임으로 변환
distributor_df = pd.DataFrame(distributor_infos,
columns = ['distributor'])
distributor_df
```

```python
# csv 파일 저장
distributor_df.to_csv("distributor.csv", encoding = "utf-8-sig")
```
